# Exploratory Data Analysis on Terrorism

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_set=pd.read_csv("terrorism.csv")
```

```python
data_set.rename(columns={'iyear':'Year','imonth':'Month','iday':"Day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'Latitude',
    'longitude':'Longitude','summary':'Summary','attacktype1_txt':'attackType','targtype1_txt':'targetType','weaptype1_txt':'Weapon','nkill':'Kill',
     'nwound':'Wound'},inplace=True)
```

```python
data_set = data_set[['Year','Month','Day','Country','State','Region','City','Latitude','Longitude',"attackType",'Kill',
               'Wound','target1','Summary','Group','targetType','Weapon','motive']]
```

```python
data_set.head()
```

```python
data_set.isnull().sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
sns.heatmap(data_set.corr(), annot=True, linewidth=0.5, fmt= '.3f')
```

```python
year = data_set['Year'].unique()
years_count = data_set['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (15,10))
sns.barplot(x = year, y = years_count)
plt.xticks(rotation = 90)
plt.xlabel('Attacking Year')
plt.ylabel('Number of Attacks Each Year')
plt.title('Attacks In Years')
plt.show()
```

```python
year_s=data_set[["Year",'Kill']].groupby("Year").sum().sort_values (by = 'Kill',ascending=False)
year_s.head(10)
```

```python
plt.subplots()
sns.barplot(data_set['Year'].value_counts()[:10].index,data_set['Kill'].value_counts()[:10].values)
plt.title('No. of Persons killed in each Years')
plt.xlabel('Year')
plt.ylabel('No. of Person Killed')
plt.xticks(rotation = 90)
plt.show()
```

```python
country = pd.DataFrame(data_set["Country"].value_counts())
print(" Name of the Countries (most affected by the attacks) : \n")
country.head(10)
```

```python
plt.subplots()
sns.barplot(data_set['Country'].value_counts()[:10].index,data_set['Country'].value_counts()[:10].values)
plt.title('Top Countries Affected')
plt.xlabel('Countries Name')
plt.ylabel('No. of Attacks(value_Count)')
plt.xticks(rotation = 90)
plt.show()
```

```python
region = pd.DataFrame(data_set["Region"].value_counts())
print("Regions affected most by the attacks: \n")
region.head()
```

```python
plt.subplots()
sns.barplot(data_set['Region'].value_counts()[:10].index,data_set['Region'].value_counts()[:10].values)
plt.title('Terrorist Activities in Region')
plt.ylabel('Number of Attacks')
plt.xlabel("Region")
plt.xticks(rotation = 90)
plt.show()
```

```python
data_set=data_set[['Group','Country','Kill']]
data_set=data_set.groupby(['Group','Country'],axis=0).sum().sort_values('Kill',ascending=False).drop('Unknown').reset_index().head(10)
data_set
```

```python
kill = data_set.loc[:,'Kill']
print('No of people killed :', int(sum(kill.dropna())))
```

```python
typeKill = data_set.pivot_table(columns='attackType', values='Kill', aggfunc='sum')
typeKill
```

```python
countryKill = data_set.pivot_table(columns='Country', values='Kill', aggfunc='sum')
countryKill
```

```python
plt.subplots(figsize=(20,10))
sns.countplot(data_set["targetType"],order=data_set['targetType'].value_counts().index,palette="gist_heat",edgecolor=sns.color_palette("mako"));
plt.xticks(rotation=90)
plt.xlabel("Attacktype",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.title("Attack per year",fontsize=20)
plt.show()
```

```python
data_set[['Group','Kill']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('Kill',ascending=False).head(10).plot(kind='bar',color='pink',figsize=(20,10))
plt.title("Top 10 terrorist group attack",fontsize=20)
plt.xlabel("terrorist group name",fontsize=15)
plt.ylabel("No of killed people",fontsize=15)
plt.show()
```
