# Image to Pencil Sketch with Python

```python
import cv2
from cv2 import imshow
import matplotlib.pyplot as plt
image = cv2.imread("Sample Image.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#plt.figure(figsize = (8,8))
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize = (8,8))
plt.imshow(gray_image, cmap = "gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

```python
inverted_image = cv2.bitwise_not(gray_image)
plt.figure(figsize = (8,8))
plt.imshow(inverted_image, cmap = "gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
image_smoothing = cv2.GaussianBlur(inverted_image, (21,21),sigmaX = 0, sigmaY = 0)
plt.figure(figsize = (8,8))
plt.imshow(image_smoothing, cmap = "gray")
plt.axis("off")
plt.title("Image_Smoothing")
plt.show()
```

```python
pencil_sketch = cv2.divide(gray_image, 255 - image_smoothing ,scale = 255)
plt.figure(figsize = (8,8))
plt.imshow(pencil_sketch, cmap = "gray")
plt.axis("off")
plt.title("Final Pencil Sketch")
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.imshow(image)
plt.axis("off")
plt.title("Original Image",size=16)
plt.figure(figsize = (10,10))
plt.imshow(pencil_sketch, cmap = "gray")
plt.axis("off")
plt.title("Final Pencil Sketch",size=16)
plt.show()
```
